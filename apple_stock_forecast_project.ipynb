```
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# Download Apple stock data
df = yf.download("AAPL", start="2019-01-01", end="2024-12-31")
df = df.reset_index()

# Use only Date and Close columns
df = df[["Date", "Close"]]
df = df.dropna()
```

```
# Prepare data for modeling
df['Date_ordinal'] = pd.to_datetime(df['Date']).map(pd.Timestamp.toordinal)
X = df['Date_ordinal'].values.reshape(-1, 1)
y = df['Close'].values

model = LinearRegression()
model.fit(X, y)
df['Predicted'] = model.predict(X)
```

```
# Plot actual vs predicted
plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Close'], label="Actual", linewidth=2)
plt.plot(df['Date'], df['Predicted'], label="Predicted", linestyle="--")
plt.title("Apple Stock Price & Linear Forecast (2019‚Äì2024)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
